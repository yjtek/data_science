```python
import heapq

test = [[1,4], [3,6], [2,8]]

q = test.copy()
heapq.heapify(q)
prev_min, prev_max = (None, None)
res = []
while q:
    curr_min, curr_max = heapq.heappop(q)
    
    if not res:
        prev_min, prev_max = curr_min, curr_max
        res.append((curr_min, curr_max))
        continue
    
    if (curr_min <= prev_min and curr_max >= prev_max):
        res.pop()
        res.append((curr_min, curr_max))
        prev_min, prev_max = curr_min, curr_max
    elif (curr_min >= prev_min and curr_max <= prev_max):
        continue
    else:
        prev_min, prev_max = curr_min, curr_max
        res.append((curr_min, curr_max))
    
res
```

### Board without snakes or ladders

- It is clear that, without snakes or ladders, most turns will involves getting stuck while getting to 100 from 94-99. 
- So the bulk of the attempts are in this range due to backtracking from 100

```python
import random
```

```python
res = []
for _ in range(10_000):
    curr = 0
    while curr != 100:
        curr = curr + random.randint(1,6)
        if curr > 100:
            curr = 100 - (curr - 100)
        res.append(curr)
```

```python
from collections import Counter
import pandas as pd
temp = pd.DataFrame.from_dict(dict(Counter(res)), orient='index').reset_index(drop=False)
temp.columns = ['index', 'count']
temp.head()
```
