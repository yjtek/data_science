## $\log$ OLS regression

- In OLS, we have occasions when we use the $\log$ term in our OLS specification

- There are 3 ways to do this
    - log-level: 1 unit change in $X$ changes $y$ by $100\beta \%$
    
    $$\ln(y) = \beta X + \epsilon $$
    
    - level-log: 1% change in $X$ changes $y$ by $\beta/100 \%$
    
    $$y = \beta \ln(X) + \epsilon $$
    
    - log-log: 1% change in $X$ changes $y$ by $\beta \%$. This is also known as the elasticity
    
    $$\ln(y) = \beta \ln(X) + \epsilon $$
- In Log-Level, taking the derivative of boths sides and rearranging:
    - $\partial X$ is the unit change in $X$. 
    - $\frac{\partial y}{y}$ is the percentage change in $y$
    - So when $\partial X = 1$ (i.e. X changes by 1 unit), then $\frac{\partial y}{y} = \beta$
    - Therefore, $\beta$ is the percentage change in $y$ when $X$ changes by 1 unit

$$\begin{aligned}
    \ln(y) &= \beta X + \epsilon \\
    \frac{\partial \ln(y)}{\partial X} &= \beta \\
    \frac{1}{y} \frac{\partial y}{\partial X} &= \beta \\
    \frac{\partial y}{y} &= \beta \cdot \partial X
\end{aligned}$$
- In level-log
    - $\partial X$ is the unit change in $X$. 
    - $\frac{\partial y}{y}$ is the percentage change in $y$
    - $\frac{\partial X}{X}$ is the proportion of change in $X$, and we multiply this by 100 to get the percentage change
    - Therefore, $\frac{\beta}{100}$ is the change in $y$ when $X$ changes by 1%

$$\begin{aligned}
    y &= \beta \ln(X) + \epsilon \\
    \frac{\partial y}{\partial X} &= \beta \frac{1}{X} \\
    \partial y &= \beta \frac{\partial X}{X} \\
    \partial y &= \beta \frac{\partial X}{X} * 100 * \frac{1}{100} \\
    \partial y &= \frac{\beta}{100} \cdot \% \Delta X \\
\end{aligned}$$ 
- In log-log
    - Differentiating and rearranging shows us that $\beta$ is the % change in $y$ following a % change in $X$

$$\begin{aligned}
    \ln(y) &= \beta \ln(X) + \epsilon \\
    \frac{1}{y} \frac{\partial y}{\partial X} &= \beta \frac{1}{X} \\
    \frac{\partial y}{y} &= \beta \frac{\partial X}{X} \\
    \% \Delta y &= \beta \cdot \% \Delta X
\end{aligned}$$ 

### The approximation hidden in log OLS

- It is **EXTREMELY IMPORTANT** to note that the coefficient interpretations above are an approximation to different degrees. We'll illustrate below why this is the case, and how sensitive each specification is to the approximation
- Log-Level
    - We can rewrite our log-level specification such that $y$ is in levels instead of log
    - Then, taking a unit change in X, we find that true percentage change of $y$ from a unit change of X is expressed as $e^{\beta \Delta X} - 1$
    - From the Taylor series, we know that $e^{\beta} = 1 + \beta + \frac{\beta^2}{2} + ... $
    - Therefore, interpreting $100 \cdot \beta$ as % change in $y$ for a unit change in $X$, is an approximation that is only valid for if $\beta \Delta X$ is small

$$\begin{aligned}
    \ln(y) &= \beta X + \epsilon \\ \\
    
    y_0 &= e^{\beta X} \\ 
    y_1 &= e^{\beta (X + \Delta X)} \\
    &= e^{\beta X} e^{\beta \Delta X} \\ \\

    \frac{y_1 - y_0}{y_0} &= \frac{e^{\beta X} e^{\beta \Delta X} - e^{\beta X}}{e^{\beta X}} \\
    &= e^{\beta \Delta X} - 1 \\
    &= 1 + \beta \Delta X + \frac{(\beta \Delta X) ^2}{2} + ... - 1 & \text{By Taylor Series} \\
    &\approx \beta \Delta X
\end{aligned}$$


- What happens when $\beta \Delta X$ is not small?
    - Let's imagine if $\Delta X = 5$ and $\beta = 1$. Then mathematically, $e^{\beta \Delta X} - 1 = 147.41$. So $\% \Delta y = 14741\% $
    - If we assume $\beta$ is the percentage change in $y$ for a unit change in $X$, then the coefficient will lead us to the conclusion that $y$ will change by $100 * 1 * 5 = 500\%$
    - This is a clear violation of the approximation from reality
- Level-Log
    - In the level log case, similar approximation takes place. 
    - The interpretation of $\beta$ as the change in $y$ when $X$ changes by 1% is only legit as an approximation 

$$\begin{aligned}
    y &= \beta \ln(X) + \epsilon \\
    \frac{\partial y}{\partial X} &= \beta \frac{1}{X} \\ 
    \partial y &= \beta \cdot \frac{\partial X}{X} \\ \\

    \text{Let } X_1 &= X_0 \cdot (1 + g) \\
    y_0 &= \beta \ln(X_0) + \epsilon \\    
    y_1 &= \beta \ln(X_1) + \epsilon \\ \\ 

    \Delta y &= y_1 - y_0 \\
    &= \beta [\ln(X_1) - \ln(X_0)] \\
    &= \beta \ln(\frac{X_1}{X_0}) \\
    &= \beta \cdot \ln (1 + g) \\ \\

    \ln (1 + g) &= g - \frac{g^2}{2} + ... \\
    &\approx g \\ \\

    \therefore \Delta y &= \beta \cdot \ln (1 + g) \\
    &= \beta g 

\end{aligned}$$

- What happens when $g$ is large? That is, $X$ changes by a large percent
    - Suppose $g = 5$ and $\beta = 1$
    - Then $\Delta y = y_1 - y_0 = \beta \cdot \ln (1 + g) = 1 * \ln(6) = 1.792$
    - But $\beta g = 5$
- Log-Log
    - The log log specification proof just combines both of the scenarios above

$$\begin{aligned}
    \ln(y) &= \beta \ln(X) \\ 
    y &= e^{\beta \ln(X)} \\
    &= e^{\ln(X^\beta)} \\
    &= x^{\beta} \\ \\

    X_1 &= X_0 (1 + g) \\ \\

    \therefore y_0 &= X_0^{\beta} \\
    y_1 &= X_1^{\beta} \\
    &= X_0^{\beta} (1 + g)^\beta \\ \\

    \therefore \frac{y_1 - y_0}{y_0} &= \frac{X_0^{\beta} ((1 + g)^\beta - 1)}{X_0} \\
    &= (1 + g)^\beta - 1 \\ \\ 

    \ln(1 + g) &= g - \frac{g^2}{2} + ... & \text{Taylor series} \\
    &\approx g \\ \\

    \therefore \frac{y_1 - y_0}{y_0} &= (1 + g)^\beta - 1 \\
    &= e^{\beta \ln (1+g)} \\
    &\approx e^{\beta g} \\ \\

    e^{\beta g} &= 1 + \beta g + \frac{(\beta g)^2}{2} + ... \\
    &\approx 1 + \beta g & \text{Taylor series} \\ \\

    \therefore \frac{y_1 - y_0}{y_0} &= (1 + g)^\beta - 1 \\
    &= \beta g
\end{aligned}$$

- What happens when $g$ is large? 
    - Suppose $g = 5$ and $\beta = 1$
    - Then the true change is $(1 + 5)^1 - 1 = 5$
    - But it is also true that $\beta g = 5$
    - Amazingly, no approximation is needed for the log-log case!!
